```python
from nba_api.stats.static import players
from nba_api.stats.endpoints import shotchartdetail
import pandas as pd
import matplotlib
import matplotlib.patches as patches
from matplotlib import pyplot as plt
from matplotlib import image as img
from matplotlib.patches import Rectangle
```

```python
player_dictionary = players.get_players()
jokic = [player for player in player_dictionary if player['full_name'] == 'Nikola Jokic']
embiid = [player for player in player_dictionary if player['full_name'] == 'Joel Embiid']
giannis = [player for player in player_dictionary if player['full_name'] == 'Giannis Antetokounmpo']
```

```python
#Get Shot Data
shots_jokic = shotchartdetail.ShotChartDetail(team_id = 0, player_id = '203999', context_measure_simple = 'FGA', season_type_all_star = ['Regular Season', 'Playoffs'])
shots_embiid = shotchartdetail.ShotChartDetail(team_id = 0, player_id = '203954', context_measure_simple = 'FGA', season_type_all_star = ['Regular Season', 'Playoffs'])
shots_giannis = shotchartdetail.ShotChartDetail(team_id = 0, player_id = '203507', context_measure_simple = 'FGA', season_type_all_star = ['Regular Season', 'Playoffs'])
```

```python
#Store in a dataframe
jokic_df = shots_jokic.get_data_frames()[0]
embiid_df = shots_embiid.get_data_frames()[0]
giannis_df = shots_giannis.get_data_frames()[0]
```

```python
#Create csv files of the three players' shot data
jokic = jokic_df.to_csv('Nikola_Jokic.csv')
embiid = embiid_df.to_csv('Joel_Embiid.csv')
giannis = giannis_df.to_csv('Giannis_Antetokounmpo.csv')
```

```python
#We do not want to count misses in the final shot chart, so I'm removing all the rows with misses here.
for index, row in jokic_df.iterrows():
    if(row[20]==0):
        jokic_df.drop(labels=[index], axis=0, inplace=True)
```

```python
for index, row in embiid_df.iterrows():
    if(row[20]==0):
        embiid_df.drop(labels=[index], axis=0, inplace=True)
```

```python
for index, row in giannis_df.iterrows():
    if(row[20]==0):
        giannis_df.drop(labels=[index], axis=0, inplace=True)
```

```python
#Function to draw the court
def draw_court(ax, color, dataframe):
    ax.add_patch(Rectangle((-250, 500), 1000, -1000,
             edgecolor = 'none',
             facecolor = 'ivory',
             fill=True,
             lw=5))
    plt.hexbin(dataframe['LOC_X'], dataframe['LOC_Y'] + 60, gridsize=(30, 30), extent=(-300, 300, 0, 940), bins='log', cmap='Greens')
    ax.plot([-220, -220], [0, 140], linewidth=2, color=color)
    ax.plot([220, 220], [0, 140], linewidth=2, color=color)
    ax.add_artist(matplotlib.patches.Arc((0, 140), 440, 315, theta1=0, theta2=180, facecolor='none', edgecolor=color, lw=2))
    ax.plot([-80, -80], [0, 190], linewidth=2, color=color)
    ax.plot([80, 80], [0, 190], linewidth=2, color=color)
    ax.plot([-60, -60], [0, 190], linewidth=2, color=color)
    ax.plot([60, 60], [0, 190], linewidth=2, color=color)
    ax.plot([-80, 80], [190, 190], linewidth=2, color=color)
    ax.add_artist(matplotlib.patches.Circle((0, 190), 60, facecolor='none', edgecolor=color, lw=2))   
    ax.add_artist(matplotlib.patches.Circle((0, 60), 15, facecolor='none', edgecolor=color, lw=2))
    ax.plot([-30, 30], [40, 40], linewidth=2, color=color)     
    ax.set_xlim(-250, 250)
    ax.set_ylim(0, 470)
    ax.set_xticks([])
    ax.set_yticks([])
```

```python
#Function to create the final shot chart
def shot_chart(df, name):
    fig = plt.figure(figsize=(4, 3.76))
    ax = draw_court(fig.add_axes([0, 0, 1, 1]), 'black', df)
    #plt.text(0, 1.05, 'Nikola Jokic\n2022-23 Regular Season', ha='right', va='top')
    matplotlib.rcParams['font.family'] = 'DejaVu Sans'
    matplotlib.rcParams['font.size'] = 10
    plt.text(-125, 400, (name +  ' 2022-23 Season'))
    plt.plot()
    plt.savefig(name + ' Shot Chart.png')

shot_chart(jokic_df, 'Nikola Jokic')
```
